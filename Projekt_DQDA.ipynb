```python
import pandas as pd

# Daten einlesen
file_path = "C:/Users/monaa/Desktop/Seminare & Vorlesungen/Projektmanagement ILV/imdb-videogames.csv"
df = pd.read_csv(file_path)

# Liste der Genre-Spalten
genre_columns = ['Action', 'Adventure', 'Comedy', 'Crime', 'Family', 'Fantasy', 'Mystery', 'Sci-Fi', 'Thriller']

# Schmelzen des DataFrames, um eine 'Genre'-Spalte zu erstellen
melted_df = df.melt(id_vars='year', value_vars=genre_columns, var_name='Genre', value_name='Present')

# Beh채lt  nur Zeilen bei, in denen das Genre vorhanden ist (True)
melted_df = melted_df[melted_df['Present'] == True]

# Gruppiert  Daten nach Jahr und Genre und z채hlt die H채ufigkeit
genre_counts_by_year = melted_df.groupby(['year', 'Genre']).size().reset_index(name='Count')

# Findet f체r jedes Jahr das Genre mit der h철chsten Anzahl
top_genre_by_year = genre_counts_by_year.loc[genre_counts_by_year.groupby('year')['Count'].idxmax()]

# Erstellt ein DataFrame, das nur das Jahr, das Genre und die Anzahl enth채lt
result_df = top_genre_by_year[['year', 'Genre', 'Count']]

# Ergebnis anzeigen
print(result_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten einlesen
file_path = "C:/Users/monaa/Desktop/Seminare & Vorlesungen/Projektmanagement ILV/imdb-videogames.csv"
df = pd.read_csv(file_path)

#  'rating' und 'votes' als Zahlen formatieren
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['votes'] = pd.to_numeric(df['votes'].str.replace(',', ''), errors='coerce')

# Bereinigen Sie die Daten, um Zeilen mit fehlenden Werten zu entfernen
df.dropna(subset=['year', 'rating', 'votes'], inplace=True)

# year in integer konventieren
df['year'] = df['year'].astype(int)

# Gruppieren der Daten nach Erscheinungsjahr und Berechnen der durchschnittlichen Bewertung und Stimmenanzahl
yearly_rating_votes = df.groupby('year')['rating', 'votes'].mean().reset_index()

# Trends f체r Bewertungen und Stimmenanzahl im Laufe der Jahre visualisieren
fig, ax1 = plt.subplots(figsize=(14, 7))

ax2 = ax1.twinx()
ax1.plot(yearly_rating_votes['year'], yearly_rating_votes['rating'], 'g-')
ax2.plot(yearly_rating_votes['year'], yearly_rating_votes['votes'], 'b-')

ax1.set_xlabel('Erscheinungsjahr')
ax1.set_ylabel('Durchschnittliche Bewertung', color='g')
ax2.set_ylabel('Durchschnittliche Stimmenanzahl', color='b')

plt.title('Einfluss des Erscheinungsjahres auf die Bewertung und Popularit채t von Spielen')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten einlesen
file_path = "C:/Users/monaa/Desktop/Seminare & Vorlesungen/Projektmanagement ILV/imdb-videogames.csv"
df = pd.read_csv(file_path)

# Entfernen der Zeilen, in denen das Jahr ein NaN-Wert  ist
df = df.dropna(subset=['year'])

# year in integer konvertieren
df['year'] = df['year'].astype(int)

# Liste der Genrespalten
genre_columns = ['Action', 'Adventure', 'Comedy', 'Crime', 'Family', 'Fantasy', 'Mystery', 'Sci-Fi', 'Thriller']

# Erstellen eines DataFrames, dass die Anzahl der Spiele pro Genre und Jahr z채hlt
genre_by_year = df.groupby('year')[genre_columns].sum()

# Berechnung der Gesamtanzahl der Spiele pro Jahr
total_games_by_year = df.groupby('year')['name'].count()

# Normalisieren der Daten, um den Anteil der Spiele pro Genre und Jahr zu erhalten
genre_by_year_normalized = genre_by_year.div(total_games_by_year, axis=0)

# Visualisierung der Daten
plt.figure(figsize=(15, 10))
for genre in genre_columns:
    plt.plot(genre_by_year_normalized.index, genre_by_year_normalized[genre], label=genre)

plt.legend()
plt.xlabel('Erscheinungsjahr')
plt.ylabel('Anteil der Spiele pro Genre')
plt.title('Entwicklung der Genrepr채ferenzen im Laufe der Zeit')
plt.show()
```

```python
import pandas as pd

# Daten einlesen
file_path = "C:/Users/monaa/Desktop/Seminare & Vorlesungen/Projektmanagement ILV/imdb-videogames.csv"
df = pd.read_csv(file_path)

# Liste der Genre-Spalten
genre_columns = ['Action', 'Adventure', 'Comedy', 'Crime', 'Family', 'Fantasy', 'Mystery', 'Sci-Fi', 'Thriller']

# Erstellen eines DataFrames, dass jedes Genre und jede Altersfreigabe z채hlt
genre_certificate_df = pd.DataFrame()

for genre in genre_columns:
    # Z채hlen, wie oft jede Altersfreigabe pro Genre vorkommt
    genre_data = df[df[genre] == True]['certificate'].value_counts().rename_axis('certificate').reset_index(name=f'{genre}_count')
    # Mergen der aktuellen Genre-Daten in den genre_certificate_df
    if genre_certificate_df.empty:
        genre_certificate_df = genre_data
    else:
        genre_certificate_df = genre_certificate_df.merge(genre_data, on='certificate', how='outer')

# F체llen der NaN-Werte mit 0 aus, da NaN bedeutet, dass es keine Spiele dieses Genres mit dieser Altersfreigabe gibt
genre_certificate_df = genre_certificate_df.fillna(0)

# Konvertieren der die Z채hlungen in Ganzzahlen
genre_certificate_df.loc[:, [f'{genre}_count' for genre in genre_columns]] = genre_certificate_df.loc[:, [f'{genre}_count' for genre in genre_columns]].astype(int)

# Ergebnis ausgeben
print(genre_certificate_df)
```
